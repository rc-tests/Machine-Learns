# *Test Models for accuracy*

```
import sklearn
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```
iris = load_iris()

df = pandas.DataFrame(iris.data, columns=iris.feature_names)
#store tabular iris in a var
df['species'] = iris.target
#assign species part of iris as target or item to be predicted and store in same var
df
```

```
x = df.drop('species', axis=1)
y = df['species']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=30)
#Split training and testing dataset for use by all models

dataset = pandas.DataFrame(x_test, columns = iris.feature_names)
#x has no species column any more since it is assigned as target or y
dataset
```

*Linear Regression Model*

```
from sklearn.linear_model import LinearRegression

myModelLR = LinearRegression()
#Model is declared as myModelLR
#LinearRegression method is assigned to it
```

```
myModelLR.fit(x, y)
#Previously declared data is fed to myModel
#Model will learn bout y by using x using Linear Regression algorithm
LRpred = myModelLR.predict(x_test)
#y is asked to be predicted using x
```

```
resLR = x_test.copy()
resLR['Prediction'] = [iris.target_names[int(i)] for i in LRpred]
resLR['Actual'] = [iris.target_names[int(i)] for i in y_test]

print(resLR)
```

```
resLR = x_test.copy()
#call for x to declare that whatever is to be shown must enter res table and
#not overwrite data in x as it would in case of x['Prediction']=LRpred
resLR['Prediction'] = LRpred
#store predicted y
resLR['Actual'] = y_test
#store actual y in the same table
resLR['Difference'] = resLR['Prediction'] - resLR['Actual']
#predicted val and actual val comparision
print(resLR)
```

```
plt.scatter(range(len(resLR)),resLR['Difference'])
plt.title('Prediction Error (Predicted - Actual)')
plt.xlabel('Test Instance Index')
plt.ylabel('Difference (Prediction - True)')
plt.axhline(0, color='red', linestyle='--')

ErrorPlotLR = plt.show()
```

```
errors = (resLR['Prediction'] != resLR['Actual']).sum()

total = len(resLR)

# Error percentage
error_percentage = (errors / total) * 100
print(f"Error Percentage: {error_percentage:.4f}%")
```

```
accuracyLR = myModelLR.score(x_test, y_test)
print(f"Accuracy: {accuracyLR * 100:.2f}%")
```

```
y_test
```

*K-Nearest Neighbors Model*

```
from sklearn.neighbors import KNeighborsClassifier

myModelKN = KNeighborsClassifier()
#New model that learns using K-Nearest Neighbors algorithm

myModelKN.fit(x,y)
myModelKN.predict(x_test)
```

```
KNpred = myModelKN.predict(x_test)

resKN = x_test.copy()
resKN['Prediction'] = KNpred
resKN['Actual'] = y_test
resKN['Difference'] = resKN['Prediction'] - resKN['Actual']
print(resKN)
```

```
plt.scatter(range(len(resKN)),resKN['Difference'])
plt.title('Prediction Error (Predicted - Actual)')
plt.xlabel('Test Instance Index')
plt.ylabel('Difference (Prediction - True)')
plt.axhline(0, color='red', linestyle='--')

ErrorKN = plt.show()
```

```
errors = (resKN['Prediction'] != resKN['Actual']).sum()

total = len(resKN)
ErrorRateKN = (errors / total) * 100
print(f"Error Percentage: {ErrorRateKN:.4f}%")
```

```
accuracyKN = myModelKN.score(x_test, y_test)
print(f"Accuracy: {accuracyKN * 100:.2f}%")
```

```
cmKN = confusion_matrix(y_test, KNpred)

disp = ConfusionMatrixDisplay(confusion_matrix=cmKN, display_labels=iris.target_names)
disp.plot(cmap="Blues")
```

*Decision Tree Classifier Model*

```
from sklearn.tree import DecisionTreeClassifier

myModelDT = DecisionTreeClassifier()
myModelDT.fit(x, y)
```

```
DTpred = myModelDT.predict(x_test)

resDT = x_test.copy()
resDT['Prediction'] = DTpred
resDT['Actual'] = y_test
resDT['Difference'] = resDT['Prediction'] - resDT['Actual']
print(resDT)
```

```
plt.scatter(range(len(resDT)),resDT['Difference'])
plt.title('Prediction Error (Predicted - Actual)')
plt.xlabel('Test Instance Index')
plt.ylabel('Difference (Prediction - True)')
plt.axhline(0, color='red', linestyle='--')

ErrorDT = plt.show()
```

```
cmDT = confusion_matrix(y_test, DTpred)

disp = ConfusionMatrixDisplay(confusion_matrix=cmDT, display_labels=iris.target_names)
disp.plot(cmap="Blues")
```

```
errors = (resDT['Prediction'] != resDT['Actual']).sum()

total = len(resDT)
ErrorRateDT = (errors / total) * 100
print(f"Error Percentage: {ErrorRateDT:.4f}%")
```

```
accuracyDT = myModelDT.score(x_test, y_test)
print(f"Accuracy: {accuracyDT * 100:.2f}%")
```
